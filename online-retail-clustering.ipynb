# Feature Engineering

Recap: the features that we want to analyze our customers are the features of: 
1. Recenecy: how many days has it been since their last purchase
2. frequency: how many times they have made the purchase
3. Monetary value: the sum/ total amount a particular customer has spent

#### Step 1:
compute line of total/ a total amount spent on a particular transaction line so that we can easily aggregate that into a sum of monetary values

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
cleaned_df
#sales unit is pound
```

#### Step 2:
with the total, we are now going to aggregate our data by grouping by Customer ID because we want a unique set of Customer IDs and then computing the aggregates for the recency value, for the frequency value and the monetary value

```python
aggregated_df = cleaned_df.groupby("Customer ID", as_index=False)\
    .agg(
        MonetaryValue = ("SalesLineTotal", "sum"),
        Frequency = ("Invoice", "nunique"),
        LastInvoiceDate = ("InvoiceDate", "max")
    )
aggregated_df.head()
```

#### Step 3:
Now we need to get our Recency feature, to do that I'm going to leverage our last Invoice Date, assuming that date will help us detemine the recency. 

For live project scenarios, you would like to do it w.r.t today() but that doesn't make sense for our case because this data is old, of 2010

```python
maxInvoiceDate = aggregated_df["LastInvoiceDate"].max()
maxInvoiceDate
```

```python
# calculate Recency: days since last transaction relative to the maximum date in our dataset
aggregated_df["Recency"] = (maxInvoiceDate - aggregated_df["LastInvoiceDate"]).dt.days
aggregated_df.head()
```

### We need to deal with these outliers now. 
You maybe tempted to remove these outliers or simply reject them, but the values are important because these represents customers that spent a lot and really frequently-the most valueable customers

### Strategy to tackle outliers:
separate out. do separate analysis or separate categorization. and then use the remaining data for our k-means clustering.
so to get rid of the outliers, 1st thing we want to establish is our quartile range, so we're going to use the IQR method to separate out the outlier data.

```python
# separate out outliers in monetary feature
M_Q1 = aggregated_df["MonetaryValue"].quantile(.25)
M_Q3 = aggregated_df["MonetaryValue"].quantile(.75)
M_IQR = M_Q3 - M_Q1

monetary_outliers = aggregated_df[ \
    (aggregated_df["MonetaryValue"]>M_Q3 + 1.5 * M_IQR) | \
    (aggregated_df["MonetaryValue"]<M_Q1 - 1.5 * M_IQR) \
].copy()
monetary_outliers.describe()
```

- there are 423 outliers. reasonably big but compared to our full set of data not that big.
- the min value is 3802 pound; max is 349164. no wonder, big gap!
- repeat same process for frequency

```python
# separate out outliers in frequency feature
f_Q1 = aggregated_df["Frequency"].quantile(.25)
f_Q3 = aggregated_df["Frequency"].quantile(.75)
f_IQR = f_Q3 - f_Q1

frequency_outliers = aggregated_df[ \
    (aggregated_df["Frequency"]>f_Q3 + 1.5 * f_IQR) | \
    (aggregated_df["Frequency"]<f_Q1 - 1.5 * f_IQR) \
].copy()
frequency_outliers.describe()
```

### This is where scaling comes into the picture
scaling is a set of methods to deal with features on different scales. There are different types of scaling. In this case, I'll use standard scaling.


Standard scaling transforms the features of your data to have a mean of 0 and a standard deviation of 1, ensuring that each feature contributes equally to the analysis.

 

Where:

- z is the standardized value,
- x is the original value,
- mu is the mean of the feature,
- sigma is the standard deviation of the feature.

this should put our data into more reasonable scale. 

1 caveat: not necessarily true, but a fair enough assumption is that z follows a normal distribution.
if you're dealing with a data that really deviates from a normal distribution that may not be the best method to use, there are other methods of scaling.

### K-means clustering
number of clusters is determined by the number of centroids we have
and to determine the appropriate centroids, we want to find a point where these centers are stable.
in the beginning the centroids will move a lot, we wanna reach a point where they don't move around a lot and it has come to a point where the distance from the points in the cluster is minimized. this distance is known as intertia. it's the vector distance or the average vector distance between these points and the appropriate center allocated to the cluster.

so we'll try a whole bunch of different values of k and see what the inertia is for each of those values of k to determine what the appropriate value of k is. through elbow method.

```python
# set an upper limit of k
max_k = 12

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_scores = []
k_value = range(2, max_k + 1)

for k in k_value:
    kmeans = KMeans(n_clusters = k, random_state = 42, max_iter = 1000)
    # cluster no.s each time we run the Kmeans algorthm on top of our scaled data
    cluster_labels = kmeans.fit_predict(scaled_data_df)
# then compute the silhouette score
    sil_score = silhouette_score(scaled_data_df, cluster_labels)
    silhouette_scores.append(sil_score)
    
    inertia.append(kmeans.inertia_)

plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1)
plt.plot(k_value, inertia, marker = 'o')
plt.title('KMeans Inertia for Different values of k')
plt.xlabel('Number of Clusters K')
plt.ylabel('WCSS')
plt.xticks(k_value)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_value, silhouette_scores, marker = 'o', color = 'orange')
plt.title('Silhouette Scores of Different values of k')
plt.xlabel('Number of Clusters K')
plt.ylabel('Silhouette Scores')
plt.xticks(k_value)
plt.grid(True)

plt.show()
```

the inertia or WCSS value starts off pretty high and slowly lowers down rapidly first and then slows down at the end.
elbow method technique - determine the point which you see the elbow of the curve. 

here it's between 4 and 5, our maximum benefit comes from allocating about 4 or 5 clusters.

now to decide between 4 and 5, let's use another metric - Silhouette Score.



we can see in the Silhouette Score chart, the Silhouette score for k=4 is higher than k = 5. so, as the tie breaker I'm going to leverage the fact that the 4-cluster option has a slightly better silhouette score than the 5-cluster option meaning that 4-clusters will have less potential overlap than the 5-cluster option. so we'll go ahead and use 4.
generally, using less cluster is better, because it reduces the complexity of your further analysis

```python
kmeans = KMeans(n_clusters=4, random_state = 42, max_iter=1000)
cluster_labels = kmeans.fit_predict(scaled_data_df)
cluster_labels
```

these numbers represent the different clusters we got.

- Cluster -3 (Monetary & Frequency Outliers) DELIGHT: Characteristics: The most valuable outliers, with extreme spending and frequent purchases. They are likely your top-tier customers who require special attention. Potential Strategy: Develop VIP programs or exclusive offers to maintain their loyalty and encourage continued engagement.

```python
# union all now - to get the full set of data
full_clustering = pd.concat([non_outliers_df, outlier_cluster_df])

cluster_labels = {
    0: "RETAIN",
    1: "RE-ENGAGE",
    2: "NURTURE",
    3: "REWARD",
    -1: "PAMPER",
    -2: "UPSELL",
    -3: "DELIGHT"
}

full_clustering["cluster_labels"] = full_clustering["Cluster"].map(cluster_labels)
full_clustering
```

## Last step - Visualize in a single visual summary

we'd take clusters with their unique labels, look at the volumes/clients/customers in those clusters and represent the features like recency, frequency, monetary value of each of the customers

```python
from seaborn import barplot, lineplot

cluster_counts = full_clustering["cluster_labels"].value_counts()
feature_means = full_clustering.groupby('cluster_labels')[['Recency', 'Frequency', 'MonetaryValue']].mean()

fig, ax1 = plt.subplots(figsize = (12, 8))

sns.barplot(x = cluster_counts.index, y = cluster_counts.values, ax = ax1,\
            palette='viridis', hue=cluster_counts.index)
ax1.set_title('Cluster Distribution with Average Feature Values')

ax2 = ax1.twinx() #adds a secondary axis, axis being mirrored, to represent another value on the y-axis, 1-side shows value; other side - avaerages

sns.lineplot(data=feature_means, ax=ax2, palette='Set2', markers='o')
ax2.set_ylabel('Average Value', color='g')

plt.show()
```

This visual really summarizes all of our findings. We have each bar representing each of our diffrent clusters and showing how many customers fall into that cluster.
- NURTURE is quite big - highest value actionable item we can take out: you really need to start nurturing these customers and this is the biggest grouping of customers that you have.
- on the line graph we have the average value of each of our features. we've got a legend at the top right that tells  recency is in green, frequency is in orange & monetary value is in the light blue color.
- imp: monetary value avg is insanely high for DELIGHT cluster compared to the rest of the clusters & that's squishing all of our other features so it's really hard to tell the difference between recency and frequency of other clusters

- divide or scale the monetary value by an order of 100 to bring this down but keep the shape of the data.

```python
full_clustering["Monetary_value_per_100_pounds"] = full_clustering["MonetaryValue"]/100

feature_means = full_clustering.groupby('cluster_labels')[['Recency', 'Frequency', 'Monetary_value_per_100_pounds']].mean()

fig, ax1 = plt.subplots(figsize = (12, 8))

sns.barplot(x = cluster_counts.index, y = cluster_counts.values, ax = ax1,\
            palette='viridis', hue=cluster_counts.index)
ax1.set_title('Cluster Distribution with Average Feature Values')

ax2 = ax1.twinx() #adds a secondary axis, axis being mirrored, to represent another value on the y-axis, 1-side shows value; other side - avaerages

sns.lineplot(data=feature_means, ax=ax2, palette='Set2', markers='o')
ax2.set_ylabel('Average Value', color='g')

plt.show()
```

that's a lot more clear. after scaling the recency is doing something similar but the shape of our feature means is a lot more clear after scaling back the monetary value. Our legend is also updated
